```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("C:/Users/Анна/Documents/1t data/attestation/yellow_tripdata_2020-01.csv", low_memory=False)
```

```python
df = df[(df["trip_distance"] > 0) & (df["trip_distance"] < 250) & (df["tip_amount"] >= 0)]
fig = plt.figure()
fig, ax = plt.subplots()
fig.suptitle('Зависимость чаевых от пройденного расстояния', fontsize=14, fontweight='bold')
ax.set_xlabel('Пройденное расстояние')
ax.set_ylabel('Чаевые')

ax.scatter(x=df['trip_distance'], y=df['tip_amount']);
```

```python
df = df[(df["passenger_count"] > 0) & (df["tip_amount"] >= 0)]
fig = plt.figure()
fig, ax = plt.subplots()
fig.suptitle('Зависимость чаевых от количнства пассажиров', fontsize=14, fontweight='bold')
ax.set_xlabel('Количество пассажиров')
ax.set_ylabel('Чаевые')

ax.scatter(x=df['passenger_count'], y=df['tip_amount']);
```
